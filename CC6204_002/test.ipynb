# Definir Red Neuronal

```python
import numpy as np

class NeuralNetwork:
    def __init__(self):
        # Inicialización de pesos y sesgos
        self.weights_input_hidden = np.random.randn(2, 2)
        self.bias_hidden = np.random.randn(2)
        self.weights_hidden_output = np.random.randn(2, 1)
        self.bias_output = np.random.randn(1)
    
    def relu(self, x):
        return np.maximum(0, x)
    
    def relu_derivative(self, x):
        return np.where(x > 0, 1, 0)
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        return x * (1 - x)
    
    def forward(self, X):
        # Capa oculta
        self.hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        self.hidden_output = self.relu(self.hidden_input)
        
        # Capa de salida
        self.output_input = np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_output
        self.output = self.sigmoid(self.output_input)
        
        return self.output
    
    def backward(self, X, y, output, learning_rate):
        # Error en la capa de salida
        self.output_error = y - output.reshape(-1)
        self.output_delta = self.output_error * self.sigmoid_derivative(output.reshape(-1))
        
        # Error en la capa oculta
        self.hidden_error = np.dot(self.output_delta.reshape(-1, 1), self.weights_hidden_output.T)
        self.hidden_delta = self.hidden_error * self.relu_derivative(self.hidden_output)
        
        # Actualización de los pesos y sesgos
        self.weights_hidden_output += learning_rate * np.dot(self.hidden_output.T, self.output_delta.reshape(-1, 1))
        self.bias_output += learning_rate * np.sum(self.output_delta, axis=0)
        self.weights_input_hidden += learning_rate * np.dot(X.T, self.hidden_delta)
        self.bias_hidden += learning_rate * np.sum(self.hidden_delta, axis=0)
    
    def train(self, X, y, epochs=10000, learning_rate=0.1):
        for epoch in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output, learning_rate)
            if epoch % 1000 == 0:
                loss = np.mean(np.square(y - output.reshape(-1)))
                print(f'Epoch {epoch}, Loss: {loss}')

# Crear y entrenar la red neuronal
nn = NeuralNetwork()
nn.train(X, y)

# Evaluar la red neuronal
for x in X:
    print(f'Input: {x}, Predicted Output: {nn.forward(x)}')
```

```python
import numpy as np

# Datos de entrada y salida para el problema XOR
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

class NeuralNetwork:
    def __init__(self):
        # Inicialización de pesos y sesgos con valores más pequeños
        self.weights_input_hidden = np.random.randn(2, 2) * 0.1
        self.bias_hidden = np.random.randn(2) * 0.1
        self.weights_hidden_output = np.random.randn(2, 1) * 0.1
        self.bias_output = np.random.randn(1) * 0.1
    
    def relu(self, x):
        return np.maximum(0, x)
    
    def relu_derivative(self, x):
        return np.where(x > 0, 1, 0)
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        return x * (1 - x)
    
    def forward(self, X):
        # Capa oculta
        self.hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        self.hidden_output = self.relu(self.hidden_input)
        
        # Capa de salida
        self.output_input = np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_output
        self.output = self.sigmoid(self.output_input)
        
        return self.output
    
    def backward(self, X, y, output, learning_rate):
        # Error en la capa de salida
        self.output_error = y - output.reshape(-1)
        self.output_delta = self.output_error * self.sigmoid_derivative(output.reshape(-1))
        
        # Error en la capa oculta
        self.hidden_error = np.dot(self.output_delta.reshape(-1, 1), self.weights_hidden_output.T)
        self.hidden_delta = self.hidden_error * self.relu_derivative(self.hidden_output)
        
        # Actualización de los pesos y sesgos
        self.weights_hidden_output += learning_rate * np.dot(self.hidden_output.T, self.output_delta.reshape(-1, 1))
        self.bias_output += learning_rate * np.sum(self.output_delta, axis=0)
        self.weights_input_hidden += learning_rate * np.dot(X.T, self.hidden_delta)
        self.bias_hidden += learning_rate * np.sum(self.hidden_delta, axis=0)
    
    def train(self, X, y, epochs=10000, learning_rate=0.1):
        for epoch in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output, learning_rate)
            if epoch % 1000 == 0:
                loss = np.mean(np.square(y - output.reshape(-1)))
                print(f'Epoch {epoch}, Loss: {loss}')

# Crear y entrenar la red neuronal
nn = NeuralNetwork()
nn.train(X, y)

# Evaluar la red neuronal
for x in X:
    print(f'Input: {x}, Predicted Output: {nn.forward(x)}')
```
